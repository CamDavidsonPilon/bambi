```python
import bambi as bmb
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import pymc3 as pm
```

```python
crossed_data = pd.read_csv("bambi/tests/data/crossed_random.csv")
```

```python
# model taken from test_build_models.py

model = bmb.Model(crossed_data)
model.add(
    "threecats[b] ~ continuous + dummy",
    family="bernoulli",
    link="logit",
)
model.build("pymc3")
display(pm.model_to_graphviz(model.backend.model))
display(model.backend.model)

pm.sample_prior_predictive(model=model.backend.model)
```
